## Numpy Tutorial

NumPy (Numerical Python) is a core Python library used for efficient numerical computation. It provides a powerful data structure called the **ndarray (N-dimensional array)**, which allows storing and manipulating large amounts of data in a compact and structured way. Unlike Python lists, NumPy arrays are homogeneous, meaning all elements have the same data type, which makes them more memory-efficient and suitable for mathematical operations.

The main concept behind NumPy is **vectorization**, which enables operations to be performed on entire arrays at once instead of using loops. This leads to cleaner code and significantly faster execution. NumPy also supports broadcasting, allowing operations between arrays of different shapes without explicitly reshaping them.

NumPy is very fast due to several key reasons. First, it is implemented in low-level languages like C, so most operations run outside the Python interpreter, reducing overhead. Second, it uses contiguous memory storage, which improves cache efficiency and speeds up data access. Third, vectorized operations eliminate the need for Python loops, which are slower compared to compiled code execution. Additionally, NumPy performs optimized numerical computations using highly efficient algorithms.

It also provides a wide range of built-in functions for linear algebra, statistical analysis, random number generation, and array manipulation. Because of its speed, efficiency, and powerful features, NumPy serves as the foundation for many advanced libraries used in data science, machine learning, and scientific computing.

# 📘 4. Memory Model of NumPy

### ✅ Key Idea

NumPy arrays store data in **contiguous memory blocks**

👉 Unlike Python lists (which store references)

---

### ✅ Python List Memory

* Stores pointers to objects
* Different memory locations

```
[1, 2, 3]
→ pointer → 1
→ pointer → 2
→ pointer → 3
```

---

### ✅ NumPy Array Memory

* Stores actual values in **continuous memory**

```
[1 2 3] → stored in one block
```

---

### ✅ Benefits

* Faster access
* Better cache performance
* Less memory usage

---

### ✅ Strides Concept (Important)

* Defines how many bytes to skip to move to next element

```python
arr.strides
```

**Strides** tell NumPy:

> “How many bytes to jump in memory to move to the next element along each axis”

👉 It is part of NumPy’s **memory model**

### ✅ Output

* Returns a tuple of bytes for each dimension

# 📘 Simple Example (1D Array)

```python
import numpy as np

arr = np.array([10, 20, 30], dtype=np.int32)
print(arr.strides)
```

### ✅ Output

```
(4,)
```

### 🔍 Meaning

* `int32` = 4 bytes
* To move from:

  * 10 → 20 → 30 → jump **4 bytes**

👉 So stride = `(4,)`

# 📘 Example (2D Array)

```python
arr = np.array([[1,2,3],
                [4,5,6]], dtype=np.int32)

print(arr.strides)
```

### ✅ Output

```
(12, 4)
```

---

### 🔍 Meaning

* Each row has 3 elements → 3 × 4 bytes = 12 bytes
* So:

  * Move to next **row** → jump 12 bytes
  * Move to next **column** → jump 4 bytes

👉 So:

```
(12, 4)
(row jump, column jump)
```

### ✅ Meaning

* Default in NumPy


### ✅ Meaning

* `True` → You can modify array
* `False` → Read-only array

## 🔹 6. `WRITEBACKIFCOPY`

### ✅ Meaning

#### typecasting

syntax : `arr.astype(dtype, copy=True)`

## Structured `dtype` 

- it allows a numpy array to store multiple fields with different data types.
- Store heterogeneous data in one array
Useful for:
    - Tabular data
    - Records (name, age, marks)
    - Binary file handling

Syntax : 

`np.dtype([('field_name1', data_type1),('field_name2', data_type2),
    ...])
`

`````python
import numpy as np

dt = np.dtype([
    ('name', 'U10'),   # string (max 10 chars)
    ('age', 'i4'),     # integer (4 bytes)
    ('marks', 'f4')    # float (4 bytes)
])

arr = np.array([
    ('Person1', 21, 85.5),
    ('Person2', 22, 90.0)
], dtype=dt)

print(arr)
print(arr['name']) # accesing the data
print(arr['name'][1])
print(arr[0])
print(arr['marks'][0])
`````

| Code    | Meaning           |
| ------- | ----------------- |
| `'i4'`  | int32 (4 bytes)   |
| `'f4'`  | float32           |
| `'U10'` | string (10 chars) |

# 📘 4. View Behavior (VERY IMPORTANT 🔥)

## ✅ Key Concept

👉 **Slicing returns a VIEW, not a copy**

---

## 🔹 Example

```python
arr = np.array([10,20,30,40])

slice_arr = arr[1:3]
slice_arr[0] = 999

print(arr)
```

👉 Output:

```
[10 999 30 40]
```

---

### 🔥 Meaning

* Changing slice → changes original array
* Because both share same memory

---

## 🔹 Check with `base`

```python
print(slice_arr.base is arr)
```

👉 Output:

```
True
```

---

## 🔹 Copy to Avoid This

```python
slice_arr = arr[1:3].copy()
```

👉 Now changes won’t affect original

#### stack

#### hstack

- `np.hstack(arrays)`

### Views and Copies

- *View* -> Shares same Memory
- *Copy* -> Creates new Memory
#### View
- A view does not create new data it just references the original array.
- syntax : ` arr.view()`
- so it is faster (no new memory)
- used for performance optimization
#### Copy
- Creates a completely new array
- syntax : `arr.copy()`
- orginal data unaffected
- Needed when modifying data
- `arr.base` to check view or copy
- if base != None -> View
- if base = None -> Copy

Lets goto `flatten() & ravel()`
flatten is returns copy
ravel is returns view

`````python
data = np.array([1,2,3,4,5])
data2 = data.copy()
data2[4] = 100
print(data) # not affected
print(data.base)
print(data2.base) # None means Copy
`````

`````python
data3 = data.view()
data4 = data
data4[1]=100
print(data)
print(data2)
print(data3) # data modified
print(data4)
print(data3.base) # no None so it is View
`````
