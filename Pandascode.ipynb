# Pandas Notes (Python Data Analysis library)

Pandas is a Python Library used for :
- Data Manipulation
- Data analysis
- Handling structured data(tables, CSV, Excel....)

- import pandas as pd
- pd.Series(data, index = None, dtype = None, name = None,copy=False)
     - data -> list, array, dict, scalar
     - index -> labels for data, default 0,1,2,3,4...
     - dtype -> datatype
     - name -> name of series
     - copy -> True for copy data | False for reference original

```python
import pandas as pd
# 1D array
# used for labeled indexing
data = pd.Series([10,20,30,40,50,60,70,80],name="sample_name") # -> Series not series
print(data)
print("Here with custom labels")
data1 = pd.Series([10,20,30,40,50,60,70,80],index = ['a','b','c','d','e','f','g','h'])
print(data1)
```

```python
# Scalar value
# scalar value means In pandas, a scalar value refers to a single, indivisible data element rather than an array or sequence
s = pd.Series(5,index=['a','b','c'])
print(s)
```

```python
# Accesing data by index labl
data1['d']
```

```python
data1['c':'h'] # Slicing
```

```python
s = pd.Series([1,2,3,6,7],name="sample_name")
print(s * 2)
print(s + 10)
```

```python
print(s.index)    # get index
print(s.values)    # get values (numpy array)
print(s.dtype)     # data type
print(s.shape)     # tuple representation of dimensions
print(s.size)      # returns int representing of total number of elements.

# number of elements
```

- 2D array
- syntax : `df = pd.DataFrame(data, index=None, columns=None, dtype=None)`
    - index -> row labels
    - columns -> column names
    

```python
dict_data = pd.Series({'a':10,'b':20,'c':30,'d':40},index = [10,2,30],name="With Dict")
print(dict_data) # in this case, index should match with dict keys
```

```python
data = {'Name': ['A', 'B'], 'Age': [20, 25]}
df = pd.DataFrame(data)
print(df)
```

#### Data input/output (File Handling) CSV
- Syntax : `df = pd.read_csv(filepath, sep=',', header=0, index_col=None)`
    - filepath : file path
    - sep -> delimiter
    - header -> row for column names
    - index_col -> set index column

- `pd.read_excel(filepath,sheet_name=0)`

- `pd.read_json(filepath)`
